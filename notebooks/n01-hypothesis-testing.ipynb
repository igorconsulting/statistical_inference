## Important Terminologies

### Dependent Variable(a.k.a. outcome variable):
A dependent variable, also known as the outcome variable, response variable, or output variable, is the variable that is observed and measured to assess the effect of the independent variable(s). It is assumed to be affected by changes in the independent variable(s).

The dependent variable changes in response to changes in the independent variable. For example, in a study on the effect of study time on test scores, the test scores are the dependent variable.

The dependent variable is typically measured quantitatively, although it can also be categorical. For example, in a medical study, the dependent variable could be the number of symptoms experienced by patients.

A study can have more than one dependent variable. For example, a psychological study might measure both stress levels and happiness as dependent variables.

### Independent Variables(a.k.a. explanatory variable):
An independent variable, also known as a predictor, explanatory variable, or input variable, is the variable that is manipulated or categorized to observe its effect on a dependent variable. It is assumed to be the cause of any changes in the dependent variable.

In experimental studies, independent variables are intentionally manipulated by the researcher to observe their effect. For example, in a clinical trial, different doses of a medication might be administered to study their effects on patient recovery.

In observational studies, independent variables are naturally occurring and are not manipulated but observed for their effects. For example, the age or gender of participants can be considered as independent variables.

A study can have more than one independent variable, which are often called factors. For example, in an agricultural study, the amount of fertilizer and the type of soil might both be independent variables.

### Example: Effects of soil moisture on plant growth
 The soil moisture is the independent variable because it is the variable that is manipulated or categorized to observe its effects on another variable. Researchers can control and vary the levels of soil moisture to observe its impact on plant growth. For instance, they might set different moisture levels such as dry, moderate, and wet conditions. As the main goal is to see how different levels of soil moisture influence the growth of plants, by varying soil moisture, researchers can isolate its effect on plant growth, ensuring other factors are kept constant.
 
 Additionally, Plant Growth is the dependent variable because it is the variable that is measured and observed to see how it responds to changes in the independent variable, the soil moisture. Plant growth can be measured in various ways, such as height, biomass, leaf number, or root length. These measurements will be taken after exposing the plants to different soil moisture levels. The growth of the plants is expected to change in response to the varying levels of soil moisture. Researchers will record and analyze these changes to understand the relationship between soil moisture and plant growth.
 
 An example of good hypothesis might be that increased soil moisture positively affects plant growth up to a certain level, beyond which excessive moisture might hinder growth.
 
 - A good experiment setup: Different groups of plants are grown under controlled conditions where the only variable that changes is the soil moisture level. For example, Group A receives 10% soil moisture, Group B receives 30%, Group C receives 50%, and so on.
 
 - An observation about Control Variables: Other factors that could influence plant growth, such as light, temperature, soil type, and nutrient levels, are kept constant across all groups to ensure that any observed differences in plant growth are due to soil moisture levels.
 
 - An important consideration about Causality: By controlling other variables and only manipulating soil moisture, researchers can make a stronger case for a causal relationship between soil moisture and plant growth.

## Mathematical(maybe Physical) and Statistical Models

The mind-boggling complexities of the universe necessitate simplified models. We, as human beings, are only just barely beginning to scratch the surface of understanding the true nature of physics, biology, and other sciences. To make sense of the intricate phenomena around us, we rely on various types of models that abstract and simplify reality.

### Mathematical and Physical Models

Mathematical models use mathematical language to describe a system. These models consist of equations and formulas that explains some features in a dataset or represent relationships between variables. Physical models, on the other hand, are tangible representations of objects or systems, often used in engineering and the physical sciences.

Mathematical and physical models help us to understand complex systems by breaking them down into simpler, more manageable components, also it helps to predict the behavior of systems under various conditions and test hypotheses and validate theories. As a Data Scientist, there is another important feature of Mathematica Models: Optimize processes and design efficient systems, specially related to decision-making.

### Examples
- **Newton’s Laws of Motion**: Mathematical equations describing the relationship between the motion of an object and the forces acting on it.
- **Climate Models**: Mathematical representations of the climate system, used to predict future climate change.
- **Scale Models**: Physical replicas of structures or systems, such as architectural models or wind tunnel models for aerodynamics testing.

Mathematical and physical models possess several key characteristics. They are deterministic, meaning that given initial conditions, the model produces a unique outcome. For example, in classical mechanics, the future state of a system is entirely determined by its current state and the laws of motion. These models are also analytical, as solutions can often be derived using algebraic or calculus-based methods. Furthermore, they are reductionist, breaking down complex phenomena into simpler, more manageable components for analysis. This approach allows for a detailed understanding of the underlying mechanisms and interactions within the system.

However, these models come with limitations. They often involve simplifications because real-world systems are usually more complex than their mathematical or physical representations, leading to approximations. Additionally, models rely on assumptions that may not hold true in all situations, potentially limiting their accuracy and applicability. Finally, the deterministic nature of these models can be a drawback, as many natural and social systems exhibit stochastic behavior that deterministic models cannot capture. These limitations highlight the importance of integrating different modeling approaches to better address the complexities of real-world phenomena.

### Statistical Models

Statistical models use rules, equations, and statistical methods to represent and analyze data or to explain some features in a dataset. These models incorporate randomness and uncertainty, making them suitable for analyzing data from real-world processes that exhibit variability. One of the key characteristics of statistical models is their probabilistic nature, which allows them to describe the likelihood of different outcomes and account for the inherent randomness in data. They are also data-driven, built and validated using observed data, which enhances their applicability to practical problems. Furthermore, statistical models are flexible, capable of handling complex relationships and interactions between variables.

However, statistical models also have their limitations. Overfitting is a common issue, where a model may fit the training data very well but perform poorly on new, unseen data. This happens when the model captures noise rather than the underlying pattern. Many statistical models rely on assumptions such as normality, independence, and homoscedasticity (equal variances), which may not always hold true, potentially compromising the validity of the results. Additionally, the interpretability of complex statistical models, such as those used in machine learning, can be challenging, making it difficult to understand how predictions are made.

Despite these limitations, statistical models are robust tools for analyzing and making inferences from data. They provide a systematic framework for hypothesis testing, estimation, and prediction, allowing researchers to draw meaningful conclusions from empirical data. By carefully validating models, checking assumptions, and avoiding overfitting, statisticians can ensure the reliability and robustness of their analyses, making statistical models indispensable in both scientific research and practical applications.

## Residuals

A simple phrase that make it simple "Residuals should be small, but models should be simple and relevant!".Residuals are essential for evaluating the goodness-of-fit of the model. Analyzing residuals helps identify whether the model adequately captures the underlying data patterns or if there are systematic errors.

Mathematically, residuals are the differences between the observed values and the values predicted by the model. Therefore, for each observation $i$:

$$\text{Residual}_i = \text{Observed}_i - \text{Predicted}_i$$


```python
import numpy as np
import pandas as pd

# Set the seed for reproducibility
np.random.seed(42)

# Number of samples
n = 1000

# Simulate independent variables
H_father = np.random.normal(175, 10, n)  # Father's height in cm
H_mother = np.random.normal(165, 10, n)  # Mother's height in cm
Nutri = np.random.normal(0, 1, n)        # Nutrition score (standardized)
PhysAct = np.random.normal(0, 1, n)      # Physical activity score (standardized)
Gender = np.random.choice([0, 1], n)     # Gender (0 = Female, 1 = Male)

epsilon = np.random.normal(0, 2, n) 
x = np.random.normal(0, 2, n)
H_adult = (H_father+H_mother)/2 + np.log(np.abs(x)+1)-np.sqrt(np.abs(x)) + x + epsilon + Gender

# Create a DataFrame
data = pd.DataFrame({
    'H_adult': H_adult,
    'H_father': H_father,
    'H_mother': H_mother,
    'Nutri': Nutri,
    'PhysAct': PhysAct,
    'Gender': Gender
})

data.head()
```

```python
import statsmodels.api as sm

# Define the independent variables (X) and the dependent variable (y)
X = data[['H_father', 'H_mother', 'Nutri', 'PhysAct', 'Gender']]
y = data['H_adult']

# Add a constant to the model (intercept)
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Get the summary of the model
model_summary = model.summary()
print(model_summary)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract residuals
residuals = model.resid

# Plot residuals vs. fitted values
plt.figure(figsize=(10, 6))
sns.residplot(x=model.fittedvalues, y=residuals, lowess=True, line_kws={'color': 'red', 'lw': 1})
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residuals vs Fitted Values')
plt.show()

# Plot histogram of residuals
plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True)
plt.xlabel('Residuals')
plt.title('Histogram of Residuals')
plt.show()

# Q-Q plot of residuals
sm.qqplot(residuals, line='45')
plt.title('Q-Q Plot of Residuals')
plt.show()

# Plot residuals vs. each independent variable
independent_vars = ['H_father', 'H_mother', 'Nutri', 'PhysAct', 'Gender']
for var in independent_vars:
    plt.figure(figsize=(10, 6))
    sns.residplot(x=data[var], y=residuals, lowess=True, line_kws={'color': 'red', 'lw': 1})
    plt.xlabel(var)
    plt.ylabel('Residuals')
    plt.title(f'Residuals vs {var}')
    plt.show()
```

## Principle of Parsimony

Occam's razor, also known as Ockham's razor or Ocham's razor (Latin: novacula Occami), is a principle used for problem-solving that advises finding explanations with the fewest necessary components and it is a fundamental concept in scientific modeling and hypothesis testing. This principle is grounded in the idea that unnecessary complexity can obscure the true relationships within the data and lead to overfitting, where the model captures noise rather than the underlying signal. It is attributed to William of Ockham, an English philosopher and theologian from the 14th century. It is often summarized by the phrase "Entia non sunt multiplicanda praeter necessitatem," meaning "Entities must not be multiplied beyond necessity."

# Important Definitions

## The Population Parameter

A population parameter is a numerical value that describes a characteristic of a population. It is a fixed value, but in practice, it is usually unknown because it is often impractical or impossible to measure every individual in the population.

Estimating population parameters from sample data is the basis of inferential statistics. This allows us to make generalizations about the population. Since it is often impractical to measure the entire population, we estimate population parameters using sample statistics, for example, estimating Sample Mean, Sample Proportion, Sample Variance, Sample Median or Sample Percentiles.

## The Hypothesis-Testing Procedure

Hypothesis testing is a form of statistical inference that uses data from a sample to draw conclusions about a population parameter or a population probability distribution.

## The Null Hypothesis ($H_0$)

The null hypothesis is a statement that there is no effect or no difference, and it is what you aim to test against. It is usually the hypothesis that researchers want to disprove or nullify.

Let $\theta$ a population parameter and $\theta_0$ be some numerical value. If the null hypothesis is the claim $\theta = \theta_0$, the notation is

$$H_0 : \theta = \theta_0$$

For example, if you are testing a new drug, the null hypothesis might be: "The new drug has no effect on patients." At this case, suppose that the time recovering mean is $t_0$ in the absence of the drug. So

$$H_0 : \theta = t_0$$

## The Alternative Hypothesis ($H_1$ or $H_a$)
The alternative hypothesis is a statement that indicates the presence of an effect or a difference. It is what you want to prove.

If the null hypothesis is $H_0 : \theta = \theta_0$, the alternative hypothesis $H_a$(or $H_1$) is one of the following

$$H_a : \theta \neq \theta_0$$

$$H_a : \theta < \theta_0$$

$$H_a : \theta > \theta_0$$

The first option is known as “two-tailed”, while the latter two are known as “one tailed”.

Continuing with the drug example, the alternative hypothesis might be: "The new drug has a positive effect on patients". So,

$$H_a : \theta < t_0$$

## The Type I and Type II errors

A Type I error occurs when the null hypothesis $H_0$ is rejected when it is actually true and the probability of making a Type I error is called the significance level and denoted by $\alpha$. This error means concluding that there is an effect or difference when there is none.

Example: In a clinical trial, concluding that a new drug is effective in lowering blood pressure when it actually has no effect.

A Type II error occurs when the null hypothesis ($H_0$) is not rejected when the alternative hypothesis ($H_a$) is true and the probability of making a Type II error is denoted by $\beta$. This error means failing to detect an effect or difference that actually exists.

At the same example, in a clinical trial, concluding that a new drug is not effective in lowering blood pressure when it actually is.

## The Significance Level($\alpha$)

The significance level($\alpha$) is a threshold set based on a combination of established scientific conventions, the specific context of the research, the potential consequences of errors, and the researcher’s experience and risk tolerance. While it involves subjective judgment, it is guided by a rational evaluation of the trade-offs between Type I and Type II errors.

### Common Significance Levels

- **0.05 (5%)**: This is the most commonly used significance level in many scientific fields. It implies that there is a 5% risk of rejecting the null hypothesis when it is true.
- **0.01 (1%)**: This indicates a 1% risk of rejecting the null hypothesis when it is true, providing stronger evidence against $H_0$.
- **0.10 (10%)**: This is less stringent, indicating a 10% risk of rejecting the null hypothesis when it is true.

## How to Make a Technical Decision

When conducting hypothesis testing, making a technical decision involves evaluating the p-value in relation to the predetermined significance level $\alpha$. This step is crucial as it determines whether the null hypothesis $H_0$ should be rejected in favor of the alternative hypothesis $H_a$.

- **If the p-value ≤ $\alpha$**: Reject the null hypothesis $H_0$. This indicates that there is sufficient statistical evidence to support the alternative hypothesis $H_a$. For instance, if the significance level is set at 0.05 and the p-value obtained from the test is 0.03, we reject $H_0$. This means that the observed data is unlikely to have occurred under the assumption of the null hypothesis, suggesting that the alternative hypothesis is more plausible.

- **If the p-value > $\alpha$**: Do not reject the null hypothesis $H_0$. In this case, there is not enough statistical evidence to support the alternative hypothesis $H_a$. For example, if the p-value is 0.07 with $\alpha$ set at 0.05, we fail to reject $H_0$. This implies that the data does not provide strong enough evidence against the null hypothesis, and we cannot conclude that the alternative hypothesis is true.

### How to Draw Conclusions and Report Results

After making the decision based on the p-value, the next step is to interpret the results within the context of the research question. This involves a comprehensive report of the findings, which should include the following elements:

1. **Summary of Results**: Clearly state the outcomes of the hypothesis test. Mention the test statistic, the p-value, and the decision regarding the null hypothesis.
   - Example: "The hypothesis test resulted in a test statistic of $2.35$ and a p-value of $0.02$. Since the p-value is less than our significance level of 0.05, we reject the null hypothesis. This suggests that there is a statistically significant effect."

2. **Implications of the Results**: Discuss what the results mean in the broader context of the research. Explain how these findings contribute to the existing body of knowledge or practical applications.
   - Example: "Rejecting the null hypothesis indicates that the new teaching method has a significant impact on student performance. This finding supports the adoption of the new method in educational settings to enhance learning outcomes."

3. **Limitations**: Address any limitations of the study that could affect the interpretation of the results. This might include sample size, experimental design, potential biases, or external validity.
   - Example: "Despite the significant findings, the study is limited by a small sample size and a short duration. Future research with larger samples and longer follow-up periods is needed to confirm these results."

4. **Future Directions**: Suggest areas for further research or additional studies that could build on the findings.
   - Example: "Future studies should explore the long-term effects of the teaching method and its applicability across different age groups and educational contexts."

By thoroughly interpreting and reporting the results, researchers provide a clear and comprehensive understanding of their findings, contributing to the scientific community and informing practical applications.

## How Design the experiment Considering Hypothesis

The hypothesis-testing procedure involves using sample data to determine whether or not $H_0$ can be rejected. If $H_0$ is rejected, the statistical conclusion is that the alternative hypothesis $H_a$ is true. To design an experiment considering hypothesis testing, a possible procedure is

1. Formulate Hypotheses
2. Choose the Significance Level($\alpha$)
3. Determine the Sample Size
4. Select the Appropriate Statistical Test
5. Perform the Statistical Test
6. Make a Technical Decision
7. Draw Conclusions and Report Results

Warning! There is a tendency to use hypothesis testing methods even when they are not appropriate. Often, estimation and confidence intervals are better tools. Use hypothesis testing only when you want to test a well-defined hypothesis.

```python
import scipy.stats as stats
class Group:
    def __init__(self, average, std, n):
        self.average = average
        self.std = std
        self.n = n

```

```python
import numpy as np
import scipy.stats as stats

def calculate_t_statistic(group1, group2):
    s_1 = (group1.std**2) / group1.n
    s_2 = (group2.std**2) / group2.n
    s = s_1 + s_2
    t = (group1.average - group2.average) / np.sqrt(s)
    return t

def calculate_critical_value(alpha, df):
    """
    Calculate the critical value for a two-tailed t-test.

    Parameters:
    alpha (float): The significance level for the hypothesis test. 
                   It represents the probability of rejecting the null hypothesis 
                   when it is actually true (Type I error). Common values are 0.05, 0.01, etc.
    df (int): The degrees of freedom. In the context of a t-test, 
              this is typically the number of observations in the sample minus one.

    Returns:
    float: The critical value from the t-distribution for the given significance level and degrees of freedom.

    Explanation:
    The function uses the cumulative distribution function (ppf - percent point function) 
    of the t-distribution to find the critical value. For a two-tailed test, the alpha is divided 
    by 2 to account for both tails of the distribution. The `stats.t.ppf` function from the `scipy.stats` 
    module is used to find the critical value such that the area under the t-distribution curve to the 
    left of the critical value is equal to 1 - alpha/2.
    """
    critical_value = stats.t.ppf(1 - alpha / 2, df)  # Divide alpha by 2 for two-tailed test
    return critical_value

def two_sample_t_test(group1, group2, alpha=0.05, tail="two"):
    """
    Perform an independent two-sample t-test to determine if there is a significant difference between two groups.

    Parameters:
    group1 (Group): An instance of the Group class containing the average, standard deviation, and sample size of the first group.
    group2 (Group): An instance of the Group class containing the average, standard deviation, and sample size of the second group.
    alpha (float, optional): The significance level for the hypothesis test. Default is 0.05.
    tail (str, optional): Specifies the type of t-test to perform. "two" for a two-tailed test, "left" for a left-tailed test, 
                          "right" for a right-tailed test. Default is "two".

    Returns:
    None

    Prints:
    - The calculated t statistic.
    - The critical t value for the given significance level and degrees of freedom.
    - The result of the hypothesis test: whether to reject the null hypothesis or not.

    Explanation:
    This function calculates the t statistic using the `calculate_t_statistic` function, which takes into account the averages, 
    standard deviations, and sample sizes of the two groups. It then calculates the degrees of freedom (df) as the minimum 
    sample size of the two groups minus one. The critical t value is calculated using the `calculate_critical_value` function, 
    which uses the given alpha, df, and tail type. Finally, the function compares the t statistic with the critical t value 
    to determine if the null hypothesis should be rejected or not. The test can be two-tailed, left-tailed, or right-tailed, 
    depending on the `tail` parameter. The results are printed to the console.
    """
    # Calculate the t statistic
    t_statistic = calculate_t_statistic(group1, group2)
    print("t Statistic:", t_statistic)
    
    # Calculate the degrees of freedom
    df = min(group1.n, group2.n) - 1
    
    # Calculate the critical value
    if tail == "two":
        critical_value = calculate_critical_value(alpha, df)
        print("Critical Value:", critical_value)
        reject = abs(t_statistic) > critical_value
    elif tail == "left":
        critical_value = stats.t.ppf(alpha, df)
        print("Critical Value:", critical_value)
        reject = t_statistic < critical_value
    elif tail == "right":
        critical_value = stats.t.ppf(1 - alpha, df)
        print("Critical Value:", critical_value)
        reject = t_statistic > critical_value
    else:
        raise ValueError("Invalid value for parameter 'tail'. Use 'two', 'left', or 'right'.")

    # Interpret the results
    if reject:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")
```

```python
group_1 = Group(average=75, std=8, n=30)
group_2 = Group(average=70, std=10, n=30)
# Perform the t-test
two_sample_t_test(group_1, group_2, alpha=0.05, tail="two")
```
